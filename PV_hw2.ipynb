# Урок 2. Дискретные случайные величины. Закон распределения вероятностей. Биномиальный закон распределения. Распределение Пуассона

## 1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

Имеем дело с дискретной случайной величеной Х - попадение в цель.
вероятность попадания высока, поэтому исользуем формулу биноминального распределения (формула Бернулли)
$$P(X=k)=C_n^k\cdot p^k\cdot q^{n-k}$$
где:
p - верояность попдание в мешень с первог выстрела 0.8,
n - количестов выстрелов - 100 выстрелов,
k - искомая вероятность - попасть ровно 85 раз,
q - верояность, что спортсмен не попадет в цель, равна 0.2, (1-0.8),

Считаем: $$P(85)=C_{100}^{85}\cdot0.8^{85}\cdot 0.2^{15}=\frac{100!}{85!\cdot15!}\cdot0.8^{85}\cdot 0.2^{15}=0.048061793700746556$$




```python
import math
import numpy as np

(math.factorial(100))/(math.factorial(15)*math.factorial(85))*0.8**85*0.2**15
```

## 2. Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?

в данной задаче верояность события крайне мала, используем для рассчетов формулу Пуассона:
$$P_m\approx \frac{\lambda^m}{m!}e^{-\lambda}$$
где:

$m_1 = 0$ для первого вопроса и $m_2 = 2$ для второго

$\lambda = p\cdot n = 0.0004\cdot5000=2$

Таким образом вероятности, что ни одна лампа не сгорит:
$$P_{m_1} \approx \frac{2^0}{0!}e^{-2}=\frac{1}{1}\cdot\frac{1}{e^{2}}\approx \frac{1}{2.71^2}\approx0.135335$$

a вероятности, что сгорит только 2 лампы:
$$P_{m_2} \approx \frac{2^2}{2!}e^{-2}=\frac{4}{2}\cdot\frac{1}{e^{2}}\approx \frac{2}{2.71^2}\approx0.27067$$

## 3. Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?

Вероятность выпадения орла равна 0.5 - используем биноминальное распределение

$$P(70)=C_{144}^{70}\cdot0.5^{70}\cdot 0.5^{74}=\frac{144!}{70!\cdot74!}\cdot0.5^{70}\cdot 0.5^{74}=0.06281178$$

```python
(math.factorial(144))/(math.factorial(70)*math.factorial(74))*0.5**70*0.5**74
```

## 4. В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые? Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый?

Вероятность что все мячи белые складываются из вероятности достать из первого ящика 2 белый и из второго ящика - 2 белых, это два независимых и совместных события:
$$P(A) = P(B_1)\cdot P(B_2)$$
Вероятноть достать два белых шара из ящика равна отношению благоприятных исходов к общему количеству вариантов вытащить 2 шара из ящика.
Для первоя ящика благоприятные исходы: $$C_2^7=\frac{7!}{2!\cdot5!}=\frac{6\cdot7}{2}=\frac{42}{2}=21$$
Всего вариантов достать 2 шара для первого ящика:$$C_2^{10}=\frac{10!}{2!\cdot8!}=\frac{9\cdot10}{2}=\frac{90}{2}=45$$
отсюда: $$P(B_1)=\frac{21}{45}=0.46666$$

Для второго ящика благоприятные исходы: $$C_2^9=\frac{9!}{2!\cdot7!}=\frac{8\cdot9}{2}=\frac{72}{2}=36$$
Всего вариантов достать 2 шара для первого ящика:
$$C_2^{11}=\frac{11!}{2!\cdot9!}=\frac{10\cdot11}{2}=\frac{110}{2}=55$$
отсюда: $$P(B_2)=\frac{36}{55}=0.6545454$$
Получаем вероятность, что все мячи будут белыми:
$$P(A_1) = P(B_1)\cdot P(B_2)= 0.46666\cdot 0.6545454=0.30545$$

Для определения вероятности что при данной выборке будет только 2 белый мяча воспользуемся формулой полной вероятности, где нам предстоит сложить вероятности что из первой корзины мы достанем 2 белых, а из второй ниодного, с вероятностью достать по одному белому из каждой корзины и с вероятностью что из 1 корзины будет ни одного белого, а из второй оба белых.
Для финального расчета осталось посчитать для первой корзины:

$B_3$ - вероятность события достать один белый шар
$B_4$ - достать ни одного белого шара

для второй корзины:

$B_5$ - вероятность события достать один белый шар
$B_6$ - достать ни одного белого шара

Считаем:
$$P(B_3)=\frac{C_1^7\cdot C_1^3}{C_2^{10}}=\frac{7\cdot3}{45}=\frac{21}{45}=0.4666666666666667$$
$$P(B_4)=\frac{C_0^7\cdot C_2^3}{C_2^{10}}=\frac{1\cdot3}{45}=\frac{3}{45}=0.06666666666666667$$
$$P(B_5)=\frac{C_1^9\cdot C_1^2}{C_2^{11}}=\frac{9\cdot2}{55}=\frac{18}{55}=0.32727272727272727$$
$$P(B_6)=\frac{C_0^9\cdot C_2^2}{C_2^{11}}=\frac{1\cdot31}{55}=\frac{1}{55}=0.01818181818181818$$

Получаем вероятность, что только 2 мяча будут белыми:
$$P(A_2) = P(B_1)\cdot P(B_6)+P(B_3)\cdot P(B_5)+P(B_4)\cdot P(B_2)= 0.46666\cdot 0.0181818 + 0.466667\cdot 0.3272727+0.0666667\cdot 0.30545=0.181574242$$

для расчета вероятности, что только один мяч будет белый у нас все есть:

$$P(A_3) = P(B_3)\cdot P(B_6)+P(B_4)\cdot P(B_5)=0.466667\cdot 0.018181+0.0666667\cdot 0.32727272727272727=0.030303$$
