### 1) Из колоды в 52 карты извлекаются случайным образом 4 карты. 
###       a) Найти вероятность того, что все карты – крести. 
###       б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

Используем формулу классической вероятности:
$$P(A)=\frac{m}{n}$$

$n$ - все равновероятные события
$m$ - все элементарные события, блоприятные для А

### а) 

А - событие, когда все 4 карты одной масти
всего, возможных событий - количество вариантов взять 4 карты из колоды, можно посчитать используя формулу комбинаторики  - Сочетание, так как нам не важно в каком порядке мы получим 4 карты: $$ C_n^k=\frac {n!}{k!(n-k!)}$$
и событий, которые бы нас устроили (получение карт одной масти) можно так же посчитать Сочетанием.
Получаем:
$$P(A)=\frac{C_4^{13}}{C_4^{52}}=\frac{\frac{13!}{4!(13-4)!}}{\frac{52!}{4!(52-4!)}}=\frac{13!4!48!}{4!9!52!}=\frac{10\cdot 11\cdot 12\cdot 13}{49\cdot 50\cdot 51\cdot 52}=\frac{17160}{270725}=0,00264106$$

###  б)

А - событие, когда выбрав 4 карты, хотя бы одна из них будет туз
Если представить все множество вариантов как можно взять 4 карты, то в нем есть подмножество вариантов где будет 1,2,3 или 4 туза, и подмножество где их нет. Судя по задаче нас интресует сумма подмножеств где есть тузы, которое можно найти вычив из множества всех вариантов подмножество вариантов без тузов:
$$m = C_4^{52}- C_4^{48}$$
Получаем:
$$m = \frac{52!}{4!\cdot48!}-\frac{48!}{4!\cdot44!}=270725-\frac{45\cdot46\cdot47\cdot48}{1\cdot2\cdot3\cdot4}=270725-\frac{4669920}{24}=270725-194580=76145$$
Теперь можем посчитать вероятность искомого события:
$$P(A)=\frac{76145}{270725}=0.28126327$$

###  в)

А - событие, когда выбрав 4 карты, одна из них будет туз
используем тот же подход что и в задаче а)с той лишь разницей что количество благоприятствующих исходов будет зависить от количества возможных исходов достать за один раз один туз, а 3 оставшиеся карты чтобы были не тузами: $$m=C_4^1\cdot C_{48}^3$$
Получаем:
$$P(A)=\frac{C_4^1\cdot C_{48}^3}{C_{52}^4}=\frac{\frac{4!}{1!(4-1)!}\cdot \frac{48!}{3!(48-3)!}}{\frac{52!}{4!(52-4!)}}=\frac{4\cdot 48!4!48!}{3!45!52!}=\frac{4\cdot 46\cdot 47\cdot 48\cdot 4}{49\cdot50\cdot51\cdot52}=\frac{166-416}{270725}=0,2555508$$

(третий вариант появился, потому что с первого раза не верно понял задание)

### 2) На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

Всего вариантов нажать 3 кнопки:
$$C_{10}^{3}=\frac{10!}{3!(10-3)!}=\frac{10!}{3!\cdot7!}=\frac{8\cdot9\cdot10}{1\cdot2\cdot3}=\frac{720}{6}=120$$
благоприятный исход только один, поэтому вероятность события А 0 открфть с первой попытки:
$$P(A)=\frac{1}{120}=0,00833$$

### 3) В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

Благоприятных исходов - достать сразу только 3 окрашенных детали, посчитаем через формулу комбинаторики Сочетание - потому что нам не важен порядок этих деталей (рабочий по условиям срузу извлекает 3 детали): 
$$C_9^3=\frac{9!}{3!6!}=\frac{7\cdot8\cdot9}{1\cdot2\cdot3}=\frac{504}{6}=84$$
столько благоприятных событию А исходов, а всего возможных вариантов достать 3 детали: $$C_{15}^3=\frac{15!}{3!\cdot12!}=\frac{13\cdot14\cdot15}{1\cdot2\cdot3}=\frac{2730}{6}=455$$
Таким образом искомая вероятность обытия А: $$P(A)=\frac{84}{455}=0,184615$$

### 4) В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

вариантов купить сразу два билета из 100: $$C_{100}^2=\frac{100!}{2!\cdot98!}=\frac{99\cdot100}{2}=\frac{9900}{2}=4950$$
вариантов купить 2 выигрышных из 2 -единственный:$$C_{2}^2=\frac{2!}{2!\cdot0!}=\frac{1}{1}=1$$
Таким образом:$$P(A)=\frac{1}{4950}=0,00020202$$
