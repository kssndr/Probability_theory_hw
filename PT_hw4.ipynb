## 1. –°–ª—É—á–∞–π–Ω–∞—è –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞ A –∏–º–µ–µ—Ç —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –ø—Ä–æ–º–µ–∂—É—Ç–∫–µ (200, 800]. –ù–∞–π–¥–∏—Ç–µ –µ–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏ –¥–∏—Å–ø–µ—Ä—Å–∏—é.

–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ $M(X)$ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–π –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ–π —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –º–æ–∂–Ω–æ –≤—ã—á–∏—Å–ª–∏—Ç—å –ø–æ —Ñ–æ—Ä–º—É–ª–µ:
$$M(X) = \frac{a + b}{2}=\frac{200+800}{2}<500$$

–î–∏—Å–ø–µ—Ä—Å–∏—è $D(X)$:
$$D(X) = \frac{(b - a)^2}{12}=\frac{(800-200)^2}{12}=\frac{360000}{12}<30000$$


## 2. –û —Å–ª—É—á–∞–π–Ω–æ–π –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ–π —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω–µ B –∏–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ –µ–µ –¥–∏—Å–ø–µ—Ä—Å–∏—è —Ä–∞–≤–Ω–∞ 0.2.  –ú–æ–∂–Ω–æ –ª–∏ –Ω–∞–π—Ç–∏ –ø—Ä–∞–≤—É—é –≥—Ä–∞–Ω–∏—Ü—É –≤–µ–ª–∏—á–∏–Ω—ã B –∏ –µ–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∑–Ω–∞—è, —á—Ç–æ –ª–µ–≤–∞—è –≥—Ä–∞–Ω–∏—Ü–∞ —Ä–∞–≤–Ω–∞ 0.5? –ï—Å–ª–∏ –¥–∞, –Ω–∞–π–¥–∏—Ç–µ –µ–µ.

–î–∏—Å–ø–µ—Ä—Å–∏—è $D(X)$:
$$D(X) = \frac{(b - a)^2}{12}=\frac{(b-0.5)^2}{12}=0.2$$
$$(b-0.5)^2=0.2\dot12$$
$$b=\pm\sqrt{2.4}+0.5$$
$$b=\sqrt{2.4}+0.5=2.05$$
$$(–ø–æ—Ç–æ–º—É \ —á—Ç–æ \ —ç—Ç–æ \ –ø—Ä–∞–≤–∞—è \ –≥—Ä–∞–Ω–∏—Ü–∞)$$

$$M(X) = \frac{a + b}{2}=\frac{0.5+2.05}{2}=1.25$$
